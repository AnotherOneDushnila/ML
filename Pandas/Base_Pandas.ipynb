#### Используем датасет про Титаник:

- Pclass — класс пассажира (1 — высший, 2 — средний, 3 — низший);

- Name — имя;

- Sex — пол;

- Age — возраст;

- SibSp — количество братьев, сестер, сводных братьев, сводных сестер, супругов на борту Титаника;

- Parch — количество родителей, детей (в том числе приемных) на борту Титаника;

- Ticket — номер билета;

- Fare — плата за проезд;

- Cabin — каюта;

- Embarked — порт посадки (C — Шербур; Q — Квинстаун; S — Саутгемптон)

- Survived - пассажир выжил или нет.

В поле Age приводится количество полных лет. Для детей меньше 1 года — дробное. Если возраст не известен точно, то указано примерное значение в формате xx.5.

```python
df1 = data[['Age', 'Survived']]
df2 = data[['Cabin', 'Name']]

concat = pd.concat([df1, df2],  axis = 1) # Конкатенация по столбцам
concat.head(5)
```

```python
df_il1 = data.iloc[:5]
df_il2 = data.iloc[7:16]
row_concat = pd.concat([df_il1, df_il2], axis = 0) # Конкатенация по строкам через индексы
row_concat
```

```python
data.loc[((data['Age'] < 20) & (data['Survived'] == 1) & (data['Pclass'] == 1))] # Маски
```

```python
data1 = data[['PassengerId', 'Pclass']]
data2 = data[['Survived', 'Age']]
merge = data1.merge(data2, how = 'outer', left_on='PassengerId', right_on='Age') # Функция объединения датафреймов. 
# ВАЖНО !!! НЕЛЬЗЯ ОБЪЕДИНИТЬ ОЬЪЕКТЫ С РАЗНЫМИ ТИПАМИ ДАННЫХ !!!
merge.head(10)
```

```python
data.groupby(["Sex", "Pclass"]).agg(avg=("Age", "mean"), avg_surv=("Survived", "mean")) # Группировка и агрегация
```
